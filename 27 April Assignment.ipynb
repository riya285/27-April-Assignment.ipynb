```python
Q1. Different Types of Clustering Algorithms and Their Differences:
Hierarchical Clustering:
Approach: Divides the dataset into a tree of clusters, creating a hierarchy.
Assumption: Assumes a nested structure in the data.
K-Means Clustering:
Approach: Divides the dataset into a pre-defined number of clusters (k) based on the mean of data points.
Assumption: Assumes clusters are spherical and equally sized.
DBSCAN (Density-Based Spatial Clustering of Applications with Noise):
Approach: Identifies dense regions separated by sparser areas.
Assumption: Assumes clusters have varying shapes and densities.
Gaussian Mixture Models (GMM):
Approach: Assumes data points are generated from a mixture of several Gaussian distributions.
Assumption: Assumes clusters follow a Gaussian distribution.
Agglomerative Clustering:
Approach: Starts with individual data points as clusters and merges them until one cluster remains.
Assumption: Similar to hierarchical clustering, assumes a nested structure.


Q2. K-Means Clustering and How It Works:
Approach: Iteratively assigns data points to the nearest cluster centroid and updates centroids based on the mean of assigned points.
Working:
Randomly initialize k centroids.
Assign each data point to the nearest centroid.
Update centroids based on the mean of points assigned to each cluster.
Repeat steps 2-3 until convergence.


Q3. Advantages and Limitations of K-Means Clustering:
Advantages:
Simplicity: Easy to understand and implement.
Efficiency: Computationally efficient for large datasets.
Scalability: Scales well with the number of data points.
Limitations:
Assumption of Circular Clusters: May struggle with non-circular or varied shaped clusters.
Sensitivity to Initial Centroids: Results can vary based on initial centroid placement.
Requires Predefined K: The user must specify the number of clusters.


Q4. Determining Optimal Number of Clusters in K-Means:
Elbow Method: Plotting the sum of squared distances against the number of clusters and selecting the "elbow" point.
Silhouette Score: Measures how well-separated clusters are; higher values indicate better-defined clusters.
Gap Statistics: Compare the within-cluster sum of squares with that of a random distribution.


Q5. Applications of K-Means Clustering:
Customer Segmentation: Identifying customer groups for targeted marketing.
Image Compression: Reducing image complexity and storage space.
Anomaly Detection: Identifying unusual patterns in data.


Q6. Interpreting Output of K-Means Clustering:
Cluster Centers: Represent centroids of clusters.
Cluster Assignment: Indicates which cluster each data point belongs to.
Inertia (Within-Cluster Sum of Squares): Measures compactness of clusters.


Q7. Challenges in Implementing K-Means Clustering:
Sensitive to Initial Centroids: Results can vary based on initializations.
Choosing K: Finding the optimal number of clusters is subjective.
Handling Outliers: Sensitive to outliers that may affect cluster assignments.
```
